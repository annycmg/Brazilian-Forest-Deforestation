```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import openpyxl
path = 'C:\\Users\\anny_\\Documents\\Brazilian_Rainforest\\'
```

```python
df = pd.read_csv(path + 'Input\\inpe_brazilian_amazon_fires_1999_2019.csv')
df.to_excel(path + 'Input\\Dataset_Brazilian_Deforestation.xlsx', index=False)
df1 = df[['year', 'month', 'state', 'firespots']]
df_year_all_states = df1.copy()
```

```python
legal_amazon = (
    df_year_all_states[['state','firespots']].groupby('state',as_index=False).sum()
    .sort_values('firespots',ascending=False)['state'].values
)

plt.figure(figsize=(12,5), dpi=85)
sns.set_style("whitegrid")
sns.barplot(data=df_year_all_states, x='firespots', y='state', estimator=sum, color='green', order=legal_amazon, ci=None)    
plt.ylabel("State", size="large")
plt.title("Total Fire Outbreaks by State", size="x-large")
plt.xlabel("Total number of fire outbreaks", size="large")
```

```python
df_year_amazonas = df1.copy()
df_year_amazonas = df_year_amazonas[df_year_amazonas['state']=='AMAZONAS']
df_year_amazonas['no_firespots'] = df_year_amazonas.groupby(by=['year'])['firespots'].transform(sum)
chart_amazonas = sns.barplot(x=df_year_amazonas['year'], y=df_year_amazonas['no_firespots'])
chart_amazonas.set_xticklabels(chart_amazonas.get_xticklabels(), rotation=90)
plt.title("Number of Firespots in Amazonas State from 1999 to 2019", size="x-large")
plt.show()

# print("Number of Firespots in Amazonas State in 2005")
# df_year_amazonas = 
```

```python
legal_amazon = (
    df_year_all_states[['state','firespots']].groupby('state',as_index=False).sum()
    .sort_values('firespots',ascending=False)['state'].values
)

fig, ax = plt.subplots(3, 3, figsize=(20, 16), sharex=True)
sns.set_style("whitegrid")
ax = ax.flat
i=0

for x in legal_amazon:
    sns.barplot(data=df_year_all_states[df_year_all_states['state'] == x], x='month',
                y='firespots', ax=ax[i], color='orange')
    ax[i].set_title(x, size='large')
    ax[i].set_xticklabels(['Jan','Fev','Mar','Apr','May','Jun','Jul','Ago','Set','Oct','Nov','Dec'], fontsize='large')
    
    if x in ['PARA', 'AMAZONAS', 'RORAIMA']:
        ax[i].set_ylabel("Total number os fire outbreaks", size='large')
    else:
        ax[i].set_ylabel("")
    if x in ['RORAIMA', 'AMAPA', 'TOCANTINS']:
        ax[i].set_xlabel("Month", size='large')
    else:
        ax[i].set_xlabel("")
    i += 1
```

```python
fullmonth = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.figure(figsize=(12,5), dpi=85)
sns.set_style("whitegrid")
sns.barplot(data=df_year_all_states, x='month', y='firespots', color="orange")
plt.title("Average Fire Oubreaks by Month", size="x-large")
plt.ylabel("Total number of fire spots", size="large")
plt.xlabel("Month", size="large")
plt.xticks(ticks=df_year_all_states['month'].unique()-1, labels=fullmonth)
```

```python
int_year = list(range(1999, 2020))
plt.figure(figsize=(12,5), dpi=85)
sns.set_style("whitegrid")
sns.lineplot(data=df_year_all_states, x='year', y='firespots', estimator='sum', color="orange")
plt.title("Sum of Fire Outbreaks Detected per Year", size="x-large")
plt.ylabel("Total number of fire outbreaks", size="large")
plt.xlabel("Year", size="large")
plt.xticks(int_year, rotation=90)
plt.grid(axis='x')
```
